The Tic Tac Toe game is implemented with a 3x3 grid, player turns, and win/tie conditions.

The Higher or Lower game generates a random number for the player to guess within a specified range, with a limited number of attempts.


````java
import java.util.Scanner;
import java.util.Random;

public class GameMenu {
    private static final Scanner scanner = new Scanner(System.in);

    public static void main(String[] args) {
        displayMenu();
    }

    private static void displayMenu() {
        System.out.println("Choose a game:");
        System.out.println("1 - Rock Paper Scissors");
        System.out.println("2 - Tic Tac Toe");
        System.out.println("3 - Higher or Lower");
        System.out.println("0 - Exit");

        int choice = scanner.nextInt();

        switch (choice) {
            case 1:
                new RockPaperScissors().play();
                break;
            case 2:
                new TicTacToe().play();
                break;
            case 3:
                new HigherOrLower().play();
                break;
            case 0:
                System.out.println("Goodbye, World!");
                return;
            default:
                System.out.println("Invalid choice. Please try again.");
        }

        // Recursively display the menu until the user chooses to exit
        displayMenu();
    }
}

class RockPaperScissors {
    public void play() {
        System.out.println("Rock Paper Scissors");
        System.out.println("Type 'r' for rock, 'p' for paper, or 's' for scissors");
        Scanner scRPS = new Scanner(System.in);
        String userChoice = scRPS.nextLine().toLowerCase();
        int random = (int) (Math.random() * 3);

        String computerChoice;
        if (random == 0) {
            computerChoice = "r";
        } else if (random == 1) {
            computerChoice = "p";
        } else {
            computerChoice = "s";
        }

        System.out.println("Computer chose: " + computerChoice);

        if (userChoice.equals("r")) {
            if (computerChoice.equals("r")) {
                System.out.println("It's a tie!");
            } else if (computerChoice.equals("p")) {
                System.out.println("You lose!");
            } else {
                System.out.println("You win!");
            }
        } else if (userChoice.equals("p")) {
            if (computerChoice.equals("r")) {
                System.out.println("You win!");
            } else if (computerChoice.equals("p")) {
                System.out.println("It's a tie!");
            } else {
                System.out.println("You lose!");
            }
        } else if (userChoice.equals("s")) {
            if (computerChoice.equals("r")) {
                System.out.println("You lose!");
            } else if (computerChoice.equals("p")) {
                System.out.println("You win!");
            } else {
                System.out.println("It's a tie!");
            }
        } else {
            System.out.println("Invalid input. Please try again.");
        }

        scRPS.close();
    }
}

class TicTacToe {
    public void play() {
        System.out.println("Tic Tac Toe");
        char[][] board = {
            {' ', ' ', ' '},
            {' ', ' ', ' '},
            {' ', ' ', ' '}
        };
        boolean player1Turn = true;
        int moves = 0;

        while (true) {
            displayBoard(board);

            char currentPlayer = player1Turn ? 'X' : 'O';
            System.out.println("Player " + (player1Turn ? "1 (X)" : "2 (O)") + ", enter your move (row and column): ");
            int row = scanner.nextInt();
            int col = scanner.nextInt();

            if (row < 0 || row > 2 || col < 0 || col > 2 || board[row][col] != ' ') {
                System.out.println("Invalid move. Try again.");
                continue;
            }

            board[row][col] = currentPlayer;
            moves++;

            if (checkWin(board, currentPlayer)) {
                displayBoard(board);
                System.out.println("Player " + (player1Turn ? "1 (X)" : "2 (O)") + " wins!");
                break;
            }

            if (moves == 9) {
                displayBoard(board);
                System.out.println("It's a tie!");
                break;
            }

            player1Turn = !player1Turn;
        }
    }

    private void displayBoard(char[][] board) {
        System.out.println("-------------");
        for (int i = 0; i < 3; i++) {
            System.out.print("| ");
            for (int j = 0; j < 3; j++) {
                System.out.print(board[i][j] + " | ");
            }
            System.out.println("\n-------------");
        }
    }

    private boolean checkWin(char[][] board, char player) {
        for (int i = 0; i < 3; i++) {
            if (board[i][0] == player && board[i][1] == player && board[i][2] == player) {
                return true; // Check rows
            }
            if (board[0][i] == player && board[1][i] == player && board[2][i] == player) {
                return true; // Check columns
            }
        }
        if (board[0][0] == player && board[1][1] == player && board[2][2] == player) {
            return true; // Check diagonal (top-left to bottom-right)
        }
        if (board[0][2] == player && board[1][1] == player && board[2][0] == player) {
            return true; // Check diagonal (top-right to bottom-left)
        }
        return false;
    }
}

class HigherOrLower {
    public void play() {
        System.out.println("Higher or Lower");
        System.out.println("You have three guesses to guess the number I am thinking of between 1-8.");
        Scanner scHL = new Scanner(System.in);
        Random random = new Random();
        int target = random.nextInt(8) + 1;
        int guessesLeft = 3;

        while (guessesLeft > 0) {
            System.out.println("Enter your guess: ");
            int guess = scHL.nextInt();
            guessesLeft--;

            if (guess == target) {
                System.out.println("Congratulations! You guessed it correctly.");
                break;
            } else if (guess < target) {
                System.out.println("The number is higher. Guesses left: " + guessesLeft);
            } else {
                System.out.println("The number is lower. Guesses left: " + guessesLeft);
            }

            if (guessesLeft == 0) {
                System.out.println("Game over. The correct number was: " + target);
            }
        }

        scHL.close();
    }
}
````
